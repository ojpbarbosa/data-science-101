```python
import pandas as pd


billboard_df = pd.read_csv('data/billboard_2000_2018_spotify_lyrics.csv',
                           encoding='iso-8859-1', na_values='unknown')
billboard_df = billboard_df[['date', 'title',
                             'duration_ms', 'spotify_id', 'lyrics']]
billboard_df.dropna()
billboard_df.head()
```

```python
duration = billboard_df['duration_ms'].drop_duplicates().dropna()
duration /= 1000 * 60
```

```python
import matplotlib.pyplot as plt


plt.hist(duration, bins=20, edgecolor='k')
plt.xlabel('Song duration (minutes)')
plt.ylabel('Count')
plt.title('Song duration distribution')
plt.show()
```

#### Assignment

Using `numpy`, calculate:

- Amplitude
- Mean, mode and median
- Standard deviation
- Variance


```python
import statistics as st
import numpy as np


amplitude = max(duration) - min(duration)
mean = duration.mean()
mode = st.mode(np.round(duration, 0))
median = duration.median()
standard_deviation = duration.std()
variance = duration.var()

mode
```

```python
plt.hist(duration, bins=20, color='gainsboro', edgecolor='gray')
plt.vlines(mean, 0, 1900, colors='orangered',
           linestyles='dashed', label='Mean', linewidth=1)
plt.vlines(mode, 0, 1900, colors='limegreen',
           linestyles='dashed', label='Mode', linewidth=1)
plt.vlines(median, 0, 1900, colors='cornflowerblue',
           linestyles='dashed', label='Median', linewidth=1)
plt.legend(loc='upper right')
plt.xlabel('Song duration (minutes)')
plt.ylabel('Count')
plt.title('Song duration distribution mean, mode and median')
plt.show()
```

```python
import seaborn as sns


sns.histplot(duration, bins=20, facecolor='gainsboro',
             color='cornflowerblue', edgecolor='gray', kde=True)

plt.xlabel('Song duration (minutes)')
plt.ylabel('Count')
plt.title('Song duration normal distribution mean')
plt.show()
```

#### Chalenge

Create a **boxplot graph** with the previous data.


```python
boxplots = plt.boxplot(duration, patch_artist=True)

boxplots['boxes'][0].set(color='gray', facecolor='gainsboro', linewidth=1)

for whiskers, caps in zip(boxplots['whiskers'], boxplots['caps']):
    whiskers.set(color='gray', linewidth=1)
    caps.set(color='gray', linewidth=1)

boxplots['medians'][0].set(color='cornflowerblue', linewidth=1)

for flier in boxplots['fliers']:
    flier.set(markeredgecolor='orangered')

plt.show()
```
