```python
filtered_dataframe = dataframe.loc[:, ['location', 'date', 'total_cases']]
filtered_dataframe.head()
```

```python
filtered_dataframe.set_index('location', inplace=True)
filtered_dataframe.head()
```

```python
brazil_dataframe = dataframe[dataframe['location'] == 'Brazil']
brazil_dataframe.head()
```

```python
dataframe[(dataframe['location'] == 'Brazil') & (dataframe['total_cases'] >= 10 ** 7)].head()
# alternatively:
# brazil_dataframe[brazil_dataframe['total_cases'] >= 10 ** 7].head()
```

```python
dataframe[dataframe['iso_code'].isin(['BRA', 'USA'])].reset_index(drop=True)
```

```python
brazil_dataframe.loc[:, 'new_cases_percentage'] = brazil_dataframe['new_cases'] / brazil_dataframe['total_cases'] * 100

brazil_dataframe.head()
```

```python
import numpy as np


brazil_dataframe.loc[:, 'log_total_cases'] = np.log(brazil_dataframe['total_cases'])

brazil_dataframe.head()
```

```python
new_cases = brazil_dataframe['new_cases']
brazil_dataframe.loc[:, 'new_cases_demean'] = new_cases - new_cases.mean()

brazil_dataframe[['date', 'new_cases', 'new_cases_demean']]
```

```python
brazil_dataframe.loc[:, 'lower_location'] = brazil_dataframe['location'].str.lower()
brazil_dataframe.loc[:, 'upper_location'] = brazil_dataframe['location'].str.upper()
brazil_dataframe.head()
```

```python
columns_to_keep = ['location', 'date', 'new_cases', 'new_cases_per_million']
```

```python
br_dataframe = dataframe[dataframe['iso_code'] == 'BRA'][columns_to_keep]

br_dataframe.rename(columns={'date': 'year_month'}, inplace=True)
br_dataframe['year_month'] = br_dataframe['year_month'].str[:7]

br_dataframe.reset_index(drop=True, inplace=True)

br_dataframe.head()
```

```python
monthly_br_dataframe = br_dataframe.groupby(['location', 'year_month'], as_index=True).mean()

monthly_br_dataframe.head()
```

```python
us_dataframe = dataframe[(dataframe['iso_code'] == 'USA') & (dataframe['new_cases'] > 0)][columns_to_keep]

us_dataframe.rename(columns={'date': 'year_month'}, inplace=True)
us_dataframe['year_month'] = us_dataframe['year_month'].str[:7]

us_dataframe.reset_index(drop=True, inplace=True)

us_dataframe.head()
```

```python
monthly_us_dataframe = us_dataframe.groupby(['location', 'year_month'], as_index=True).mean()

monthly_us_dataframe.head()
```

```python
merged_monthly_br_us_dataframe = pd.merge(monthly_br_dataframe, monthly_us_dataframe, on='year_month', suffixes=('_br', '_us'))
merged_monthly_br_us_dataframe.head()
```
